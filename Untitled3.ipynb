```python
import pandas as pd

# Load the Titanic dataset from a direct URL
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)

# Display the first few rows of the dataset
titanic_df.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for the plots
sns.set(style="whitegrid")

# Visualizing the distribution of 'Pclass', 'Age', and 'Sex'
plt.figure(figsize=(15, 5))

# Plotting the distribution of Pclass
plt.subplot(1, 3, 1)
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Distribution of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')

# Plotting the distribution of Age
plt.subplot(1, 3, 2)
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')

# Plotting the distribution of Sex
plt.subplot(1, 3, 3)
sns.countplot(x='Sex', data=titanic_df)
plt.title('Distribution of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Checking for missing values in each column
print("Missing values in each column:")
titanic_df.isnull().sum()
```

```python

# Handling missing values
# Filling missing Age values with the median
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Dropping rows with missing Embarked values
titanic_df.dropna(subset=['Embarked'], inplace=True)

# Filling missing Fare values with the median (if any)
titanic_df['Fare'].fillna(titanic_df['Fare'].median(), inplace=True)

# Check again for missing values
print("\nMissing values after handling:")
titanic_df.isnull().sum()

```

```python


# Check unique values before encoding
print("\nUnique values in 'Sex':", titanic_df['Sex'].unique())
print("Unique values in 'Embarked':", titanic_df['Embarked'].unique())

# Encoding categorical variables
# Map 'Sex' to numerical values: male = 0, female = 1
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})

# Map 'Embarked' to numerical values:
# C = 0, Q = 1, S = 2
titanic_df['Embarked'] = titanic_df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

# Display the first few rows to check the encoded values
print("\nEncoded values:")
print(titanic_df[['Sex', 'Embarked']].head())

```

```python

from sklearn.preprocessing import StandardScaler

# Standardizing 'Age' and 'Fare'
scaler = StandardScaler()
titanic_df[['Age', 'Fare']] = scaler.fit_transform(titanic_df[['Age', 'Fare']])

# Display the first few rows to check the standardized values
titanic_df[['Age', 'Fare']].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# Selecting features and target variable
X = titanic_df[['Pclass', 'Sex', 'Age', 'Fare']]
y = titanic_df['Survived']

# Split the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Implementing k-Nearest Neighbors (k-NN)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Making predictions on the test set
knn_predictions = knn.predict(X_test)

# Evaluate the performance of k-NN
from sklearn.metrics import accuracy_score, classification_report

knn_accuracy = accuracy_score(y_test, knn_predictions)
knn_classification_rep = classification_report(y_test, knn_predictions)

print(f"Accuracy of k-NN: {knn_accuracy * 100:.2f}%")
print("\nClassification Report for k-NN:\n", knn_classification_rep)
```

```python
# Implementing Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)

# Making predictions on the test set
dt_predictions = decision_tree.predict(X_test)

# Evaluate the performance of Decision Tree
dt_accuracy = accuracy_score(y_test, dt_predictions)
dt_classification_rep = classification_report(y_test, dt_predictions)

print(f"Accuracy of Decision Tree: {dt_accuracy * 100:.2f}%")
print("\nClassification Report for Decision Tree:\n", dt_classification_rep)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)

# Preprocessing: Selecting relevant features and the target variable
features = ['Pclass', 'Fare', 'Survived']
titanic_df = titanic_df[features].dropna()

# Define the features (X) and target variable (y)
X = titanic_df[['Pclass', 'Fare']]
y = titanic_df['Survived']

# Split the data into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the k-NN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Train the Decision Tree model
decision_tree = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
decision_tree.fit(X_train, y_train)

# Function to visualize decision boundaries
def plot_decision_boundaries(model, X, y, title):
    # Adjusting the step size to 0.1 for faster computation
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))

    # Create a DataFrame for prediction with valid feature names
    Z = model.predict(pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=X.columns))
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(12, 6))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, edgecolors='k', marker='o', s=100)
    plt.title(title)
    plt.xlabel('Pclass')
    plt.ylabel('Fare')
    plt.show()

# Decision boundaries for k-NN
plot_decision_boundaries(knn, X, y, 'k-NN Decision Boundary')

# Decision boundaries for Decision Tree
plot_decision_boundaries(decision_tree, X, y, 'Decision Tree Decision Boundary')
```

```python
# Ensure that the predictions match the size of the test set
print(f"Length of y_test: {len(y_test)}")
print(f"Length of knn_predictions: {len(knn_predictions)}")
print(f"Length of dt_predictions: {len(dt_predictions)}")

# Re-make predictions if necessary
if len(knn_predictions) != len(y_test):
    knn_predictions = knn.predict(X_test)
if len(dt_predictions) != len(y_test):
    dt_predictions = decision_tree.predict(X_test)

# Extracting performance metrics for both models
knn_report = classification_report(y_test, knn_predictions, output_dict=True)
dt_report = classification_report(y_test, dt_predictions, output_dict=True)

# Calculating accuracy for both models
knn_accuracy = accuracy_score(y_test, knn_predictions)
dt_accuracy = accuracy_score(y_test, dt_predictions)

# Collecting metrics into a DataFrame
performance_data = {
    'Model': ['k-NN', 'Decision Tree'],
    'Accuracy': [knn_accuracy, dt_accuracy],
    'Precision': [knn_report['1']['precision'], dt_report['1']['precision']],
    'Recall': [knn_report['1']['recall'], dt_report['1']['recall']],
    'F1 Score': [knn_report['1']['f1-score'], dt_report['1']['f1-score']]
}

performance_df = pd.DataFrame(performance_data)

# Plotting the performance metrics
performance_df.set_index('Model').plot(kind='bar', figsize=(10, 6), color=['blue', 'orange', 'green', 'red'])
plt.title('Performance Comparison of k-NN and Decision Tree')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.legend(title='Metrics')
plt.grid(axis='y')
plt.show()
```
